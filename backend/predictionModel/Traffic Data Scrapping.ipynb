```python
import requests
import json

url = "https://www.waze.com/live-map/api/georss?top=32.028910650732975&bottom=31.289689658432195&left=72.76382446289064&right=74.11514282226564&env=row&types=alerts,traffic"
headers = {
    "User-Agent": "Mozilla/5.0"
}

response = requests.get(url, headers=headers)

if response.status_code == 200:
    try:
        data = response.json()  # Parse JSON response
        print(json.dumps(data, indent=4))  # Pretty-print JSON
    except json.JSONDecodeError:
        print("Response is not in JSON format. Here is the raw output:")
        print(response.text)
else:
    print(f"Error {response.status_code}: {response.reason}")
```

```python
import requests
import json

# Waze API URL
url = "https://www.waze.com/live-map/api/georss?top=32.98786041184438&bottom=30.02652224772683&left=69.97741699218751&right=75.38269042968751&env=row&types=alerts"

# Fetch the data
response = requests.get(url)

if response.status_code == 200:
    data = response.json()  # Parse JSON response
    
    # Extract alerts
    alerts = []
    for alert in data.get("alerts", []):  # Check if 'alerts' key exists
        alert_info = {
            "type": alert.get("type", "N/A"),
            "subtype": alert.get("subtype", "N/A"),
            "latitude": alert.get("location", {}).get("y", "N/A"),
            "longitude": alert.get("location", {}).get("x", "N/A"),
            "road": alert.get("roadType", "N/A"),
            "reliability": alert.get("reliability", "N/A"),
            "pubMillis": alert.get("pubMillis", "N/A"),
            "reportDescription": alert.get("reportDescription", "N/A"),
        }
        alerts.append(alert_info)

    # Convert to JSON format for readability
    json_output = json.dumps(alerts, indent=4)
    print(json_output)
else:
    print(f"Failed to fetch data. Status Code: {response.status_code}")
```

```python
import requests
import json

# Define bounding boxes for specific locations
locations = {
    "Canal Road": {"top": 31.4750, "bottom": 31.4200, "left": 73.0800, "right": 73.1500},
    "Kohinoor City": {"top": 31.4300, "bottom": 31.4100, "left": 73.0650, "right": 73.0850},
    "Susan Road": {"top": 31.4200, "bottom": 31.4000, "left": 73.0900, "right": 73.1100},
    "Ghulam Muhammad Abad": {"top": 31.4400, "bottom": 31.4100, "left": 73.0500, "right": 73.0800}
}

# Waze API base URL
base_url = "https://www.waze.com/live-map/api/georss?env=row&types=alerts"

# Fetch alerts for each location
traffic_data = {}

for area, bbox in locations.items():
    url = f"{base_url}&top={bbox['top']}&bottom={bbox['bottom']}&left={bbox['left']}&right={bbox['right']}"
    response = requests.get(url)
    
    if response.status_code == 200:
        data = response.json()
        alerts = []
        
        for alert in data.get("alerts", []):
            alert_info = {
                "type": alert.get("type", "N/A"),
                "subtype": alert.get("subtype", "N/A"),
                "latitude": alert.get("location", {}).get("y", "N/A"),
                "longitude": alert.get("location", {}).get("x", "N/A"),
                "road": alert.get("roadType", "N/A"),
                "reliability": alert.get("reliability", "N/A"),
                "pubMillis": alert.get("pubMillis", "N/A"),
                "reportDescription": alert.get("reportDescription", "N/A"),
            }
            alerts.append(alert_info)
        
        traffic_data[area] = alerts
    else:
        print(f"Failed to fetch data for {area}. Status Code: {response.status_code}")

# Save data to JSON file
with open("traffic_alerts.json", "w") as f:
    json.dump(traffic_data, f, indent=4)

print("Traffic alerts fetched successfully!")
```

```python
import json
import random
import requests
from datetime import datetime, timedelta

# Define bounding boxes for specific locations in Faisalabad
locations = {
    "Canal Road": {"lat": 31.4500, "lon": 73.1200},
    "Ghulam Muhammad Abad": {"lat": 31.4300, "lon": 73.0800},
    "Susan Road": {"lat": 31.4200, "lon": 73.1000},
    "People's Colony": {"lat": 31.4150, "lon": 73.1100},
    "Kohinoor City": {"lat": 31.4320, "lon": 73.0850},
    "Clock Tower": {"lat": 31.4180, "lon": 73.0720},
    "D Ground": {"lat": 31.4250, "lon": 73.1150},
    "Madina Town": {"lat": 31.4400, "lon": 73.1300}
}

# Waze API base URL
base_url = "https://www.waze.com/live-map/api/georss?env=row&types=alerts"

# Function to create a date within 2023
def create_date():
    start_date = datetime(2023, 1, 1)
    end_date = datetime(2023, 12, 31)
    delta = end_date - start_date
    days = random.randint(0, delta.days)
    return (start_date + timedelta(days=days)).strftime("%Y-%m-%d %H:%M:%S")

# Function to create traffic and environmental data
def create_traffic_data(area, lat, lon):
    url = f"{base_url}&lat={lat}&lon={lon}"
    response = requests.get(url)
    
    return {
        "Timestamp": create_date(),
        "Latitude": lat,
        "Longitude": lon,
        "Location": area,
        "Vehicle_Count": random.randint(5, 500),
        "NO2_Emissions": round(random.uniform(10, 150), 2),
        "Weather_Conditions": random.choice(["Clear", "Cloudy", "Rainy", "Foggy", "Hazy"]),
        "Traffic_Flow_Speed": random.randint(10, 80),  # in km/h
        "Rush_Hour_Indicator": random.choice(["Yes", "No"]),
        "Suggested_Route_Adjustment": random.choice(["None", "Use alternate road", "Avoid main road", "Delay travel"]),
        "Waze_API_URL": url,
    }

# Create data for each location
traffic_data = {area: [create_traffic_data(area, details["lat"], details["lon"]) for _ in range(random.randint(2, 5))] for area, details in locations.items()}

# Print a sample of the data
for area, alerts in list(traffic_data.items())[:3]:  # Print data for first 3 locations
    print(f"Location: {area}")
    for alert in alerts[:2]:  # Print first 2 alerts per location
        print(json.dumps(alert, indent=4))
    print("-" * 40)

# Save data to JSON file
with open("mock_traffic_alerts.json", "w") as f:
    json.dump(traffic_data, f, indent=4)

print("Traffic alerts for Faisalabad fetched successfully!")
```

```python
import json
import csv
import random
from datetime import datetime, timedelta

# Load the JSON data from the file
with open("mock_traffic_alerts.json", "r") as f:
    traffic_data = json.load(f)

# Define CSV file name
csv_file = "traffic_alerts.csv"

# Define CSV column headers
headers = [
    "Timestamp", "Latitude", "Longitude", "Location", "Vehicle_Count", "NO2_Emissions", 
    "Weather_Conditions", "Traffic_Flow_Speed", "Rush_Hour_Indicator", "Suggested_Route_Adjustment", "Waze_API_URL"
]

# Generate random date within 2023
def random_date():
    start_date = datetime(2023, 1, 1)
    end_date = datetime(2023, 12, 31)
    delta = end_date - start_date
    random_days = random.randint(0, delta.days)
    return (start_date + timedelta(days=random_days)).strftime("%Y-%m-%d %H:%M:%S")

# Write data to CSV
with open(csv_file, "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(headers)  # Write header row
    
    # Iterate through locations and alerts
    csv_rows = []
    for location, alerts in traffic_data.items():
        for alert in alerts:
            row = [
                random_date(), alert["Latitude"], alert["Longitude"], location, alert["Vehicle_Count"],
                alert["NO2_Emissions"], alert["Weather_Conditions"], alert["Traffic_Flow_Speed"], 
                alert["Rush_Hour_Indicator"], alert["Suggested_Route_Adjustment"], alert["Waze_API_URL"]
            ]
            writer.writerow(row)
            csv_rows.append(row)

print(f"Data successfully saved to {csv_file}")

# Print a few lines of the CSV data
print("Sample data from CSV:")
for row in csv_rows[:5]:
    print(row)
```
